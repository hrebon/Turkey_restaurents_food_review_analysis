# Handling missing data of Cuisines column below

```python
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return value
    else:
        return 'others'
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
```

# here identifying high-quality restaurants and their top cuisines

```python
df6 = df[['cuisines', 'total_reviews','name','rating']]
df7 = df6.groupby(['cuisines','name','rating'])['total_reviews'].sum()
df7 = df7.to_frame()
df7 = df7.sort_values('total_reviews', ascending=False)
df7
```
